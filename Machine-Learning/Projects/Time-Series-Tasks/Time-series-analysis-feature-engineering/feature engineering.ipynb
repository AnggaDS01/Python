```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
birth_df = pd.read_csv('./dataset/daily-total-female-births-CA.csv', index_col='date', parse_dates=['date'])
birth_df.head()
```

```python
birth_df = birth_df.assign(
    lag1 = birth_df['births'].shift(1),
    lag2 = birth_df['births'].shift(2),
    lag3 = birth_df['births'].shift(3),
)
```

```python
birth_df
```

```python
birth_df['births'].plot()
```

```python
birth_df['births'].rolling(10).mean().plot()
```

```python
birth_df = birth_df.assign(
    MA3 = birth_df['births'].rolling(3).mean(),
    STD4 = birth_df['births'].rolling(4).std(),
    MAX5 = birth_df['births'].rolling(5).max(),
)

birth_df
```

```python
birth_df['day'] = birth_df.index.day_name()
birth_df
```
