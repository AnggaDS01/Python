```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Types of Trends

* Deterministic Trends: These are trends that consistently increase or decrease.
* Stochastic Trends: These are trends that increase and decrease inconsistently.

```python
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m') 
    
df = pd.read_csv('./dataset/shampoo_sales.csv', index_col='Month', parse_dates=['Month'], date_parser=parser)
df.plot()
```

```python
new_df = df.copy(deep=True)
new_df['diff'] = df.diff()

new_df['diff'].plot()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(new_df['Sales'], model='multiplicative')

result.plot()
plt.show()
```

## Detrend by Model Fitting

```python
from sklearn.linear_model import LinearRegression

X = np.array([i for i in range(len(df))]).reshape(-1,1)
y = df.values

model = LinearRegression()
model.fit(X, y)

trend = model.predict(X)

plt.plot(trend)
plt.plot(y)
plt.show()

detrend = [y[i] - trend[i] for i in range(len(df))]

plt.plot(detrend)
plt.show()
```
