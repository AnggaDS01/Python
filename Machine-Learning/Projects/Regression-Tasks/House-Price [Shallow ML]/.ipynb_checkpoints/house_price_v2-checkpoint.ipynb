## Create necessary functions

```python
def diagnose_data(df):
    print(f'shape:\n{df.shape}\n')
    print(f'columns:\n{sorted(df.columns)}\n')
    print(f'n dtypes:\n{df.dtypes.value_counts()}\n')
    print(f'n uniques:')
    for col in df.columns:
        print(f'{col}: {df[col].nunique()}')
    print(f'\nunique values:')
    for col in df.columns:
        print(f'{col}: {df[col].unique()}')
    print(f'\nnull values:')
    for col in df.columns:
        print(f'{col}: {df[col].isnull().sum()}')
    print('\ninfo:')
    df.info()
    return df.head()

def get_data_types(df, types='all'):
    numeric_cols = [col for col in df.columns if df[col].dtype != object]
    cat_cols = [col for col in df.columns if train_df[col].dtype == object]
    
    if types == 'object':
        return df[cat_cols]
    elif types == 'numeric':
        return df[numeric_cols]
    elif types == 'all':
        return df
    else:
        print('Check your argument!!')
```

```python
def NaN_df(df, tresshold=.5):
    percent_missing = df.isnull().sum() / df.isnull().count()
    na_df = pd.concat(
        [
            df.isnull().sum(), 
            percent_missing
        ], 
        axis='columns', 
        keys=[
                'Total NaN', 
                'Percent NaN'
            ]
    ).sort_values(
        by='Percent NaN', 
        ascending=False)
    new_na_df = na_df[na_df['Percent NaN'] > 0]
    return new_na_df[new_na_df['Percent NaN'] > tresshold]
```

```python
numeric_df = get_data_types(train_df, 'numeric').isnull().sum()
numeric_cols_nan = numeric_df[numeric_df > 0].index

def get_best_corr_plot(df, target_col, row=1, col=2, ct='interpolate'):
    mean = [df[col].mean() for col in target_col]
    median = [df[col].median() for col in target_col]
    interpolate = [df[col].interpolate() for col in target_col]
    
    fig, ax = plt.subplots(row, col, figsize=(10, 4), dpi=100)
    for i in range(row*col):
        plt.subplot(row, col, i+1)
        plt.scatter(
            df[target_col[i]].fillna(mean[i]) if ct == 'mean' else df[target_col[i]].fillna(median[i]) if ct == 'median' else interpolate[i], 
            df['SalePrice'],
            edgecolors='white', 
            linewidths=.5,
            label=f"corr: {round(df['SalePrice'].corr( df[target_col[i]].fillna(mean[i]) if ct == 'mean' else df[target_col[i]].fillna(median[i]) if ct == 'median' else pd.Series(interpolate[i]) ), 5)}"
        )
        plt.gca().set_xlabel(target_col[i])
        plt.gca().set_ylabel("SalePrice")
        plt.legend()
    plt.subplots_adjust(
#                     left=0.1,
#                     bottom=0.1, 
#                     right=0.9, 
#                     top=0.9, 
                    wspace=0.3, 
                    hspace=0.4
                    )
    plt.show()
```

```python
from sklearn.linear_model import LassoCV
# Make a list of 50 values between 0.000001 & 1,000,000
alphas = np.logspace(-6, 6, 50)

# We will do 5-fold cross validation
reg = LassoCV(alphas=alphas, fit_intercept=False, cv=5)
reg.fit(X_train, y_train)
# y_train_pred = reg.predict(X_train)
# y_test_pred = reg.predict(X_test)

print(f"LassoCV: Chosen alpha = {reg.alpha_}")
```

```python
# n_folds equals to 5 here
n_folds = reg.mse_path_.shape[1]
# Calculate the mean and standard error for MSEs
mse_mean = reg.mse_path_.mean(axis=1)
mse_std = reg.mse_path_.std(axis=1)
# Std Error = Std Deviation / SQRT(number of samples)
mse_std_error = mse_std / np.sqrt(n_folds)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
# We multiply by 1.96 for a 95% Confidence Interval
pd.DataFrame(
    {
        'alpha': reg.alphas_,
        'Mean MSE': mse_mean,
        'Upper Bound MSE': mse_mean + 1.96 * mse_std_error,
        'Lower Bound MSE': mse_mean - 1.96 * mse_std_error,
    }
).set_index('alpha')[
        ['Mean MSE', 'Upper Bound MSE', 'Lower Bound MSE']
    ].plot(
        title='Regularization plot (MSE vs alpha)',
        marker='.', logx=True, ax=ax
)
    # Color the confidence interval
plt.fill_between(
    reg.alphas_,
    mse_mean + 1.96 * mse_std_error,
    mse_mean - 1.96 * mse_std_error,
    alpha=.2
)
# Print a vertical line for the chosen alpha
ax.axvline(reg.alpha_, linestyle='--', color='k')
ax.set_xlabel('Alpha')
ax.set_ylabel('Mean Squared Error')
```

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import make_scorer,r2_score

def test_model(model, X_train=X_train, y_train=y_train):
    cv = KFold(n_splits=7,shuffle=True,random_state=42)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model,X_train,y_train,cv=cv,scoring=r2)
    score = [r2_val_score.mean()]
    return score
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=reg.alpha_, fit_intercept=False)
lasso_model.fit(X_train, y_train)
test_model(lasso_model)
```

```python
from sklearn.linear_model import RidgeCV

# Make a list of 50 values between 0.000001 & 1,000,000
alphas_ridge = np.logspace(-6, 6, 50)

# We will do 5-fold cross validation
ridge_reg = RidgeCV(alphas=alphas_ridge, fit_intercept=False, cv=5)
ridge_reg.fit(X_train, y_train)
# y_train_pred = reg.predict(X_train)
# y_test_pred = reg.predict(X_test)

print(f"RidgeCV: Chosen alpha = {ridge_reg.alpha_}")
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=ridge_reg.alpha_, fit_intercept=False)
ridge_model.fit(X_train, y_train)
test_model(ridge_model)
```
