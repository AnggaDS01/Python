```python
list_object_null = [col for col in concat_data.columns if (concat_data[col].dtype == object) & (concat_data[col].isnull().sum() != 0)]
list_numeric_null = [col for col in concat_data.columns if (concat_data[col].dtype != object) & (concat_data[col].isnull().sum() != 0)]
```
