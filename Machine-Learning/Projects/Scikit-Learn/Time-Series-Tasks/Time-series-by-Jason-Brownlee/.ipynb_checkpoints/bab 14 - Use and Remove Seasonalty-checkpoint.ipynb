```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./dataset/daily-minimum-temperatures-in-me.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

## Seasonal Adjustment with Differencing

```python
df.info()
```

```python
df.iloc[:, 0].astype(float)
```

```python
import re

for i in df.values.flatten():
    if not re.search(r'^\d{1,}\.\d{1,}$', i, re.I):
        print(i)
```

```python
df.iloc[:, 0] = df.iloc[:, 0].apply(lambda x: re.sub(r'\?', '', x))
for i in df.values.flatten():
    if not re.search(r'^\d{1,}\.\d{1,}$', i, re.I):
        print(i)
```

```python
df.iloc[:, 0] = df.iloc[:, 0].astype(float)
df.info()
```

```python
df.plot()
```

```python
days_in_year = 365
df.diff(days_in_year).plot()
plt.show()
```

```python
monthly = df.resample('M').mean()
monthly.plot()
plt.show()
display(monthly)
```

```python
months_in_year = 12
monthly.diff(months_in_year).plot()
plt.show()
display(monthly.diff(months_in_year))
```

```python
X = df.values
diff = list()
days_in_year = 365
for i in range(days_in_year, len(X)):
    month_str = str(df.index[i].year-1)+'-'+str(df.index[i].month)
    month_mean_last_year = df[month_str].mean()
    value = X[i] - month_mean_last_year
    diff.append(value)
plt.plot(diff)
plt.show()
```

```python
month_str = str(df.index[365].year-1)+'-'+str(df.index[365].month)
df.values[0] - df.loc[month_str].mean()
```

## Seasonal Adjustment with Modeling

```python
X = [i%365 for i in range(0, len(df))]
y = df.values.flatten()
degree = 4
coef = np.polyfit(X, y, degree)

curve = []
for i in range(len(X)):
    value = coef[-1]
    for d in range(degree):
        value += X[i]**(degree-d) * coef[d]
    curve.append(value)

plt.plot(df.values)
plt.plot(curve, linewidth=4)
plt.show()
```

```python
X = [i%365 for i in range(0, len(df))]
y = df.values.flatten()
degree = 4
coef = np.polyfit(X, y, degree)

curve = []
for i in range(len(X)):
    value = coef[-1]
    for d in range(degree):
        value += X[i]**(degree-d) * coef[d]
    curve.append(value)

values = df.values
diff = []
for i in range(len(values)):
    value = values[i] - curve[i]
    diff.append(value)

plt.plot(diff)
plt.show()
```
