# Evolusi Bahasa dan Pemrosesan Bahasa Alami (NLP)

Bahasa telah menjadi bagian penting dari evolusi manusia. Dengan berkembangnya bahasa, manusia bisa berkomunikasi lebih baik, baik antar individu maupun antar kelompok. Awalnya, manusia menggunakan gambar di dinding goa untuk menyampaikan pesan. Seiring waktu, gambar-gambar ini berkembang menjadi karakter dan tulisan.

**Mengapa ini penting?** Karena dengan adanya tulisan, informasi bisa disimpan dan diteruskan dari satu generasi ke generasi berikutnya. Bayangkan jika semua pengetahuan yang kita miliki sekarang hanya ada di kepala orang-orang dan tidak pernah ditulis. Kita akan kehilangan banyak informasi berharga setiap kali seseorang meninggal.

**Kemajuan Teknologi dan Informasi**: Seiring bertambahnya informasi yang kita simpan, kita membutuhkan cara untuk mengolah dan memahami data tersebut. Dalam sepuluh tahun terakhir, banyak kemajuan telah dibuat dalam pengenalan gambar dan suara. Namun, kemajuan dalam Pemrosesan Bahasa Alami (NLP) baru terjadi belakangan ini, meskipun penelitian di bidang ini sudah berlangsung selama beberapa dekade.

**Apa itu NLP?** NLP adalah bidang yang berfokus pada bagaimana komputer bisa memahami dan memproses bahasa manusia. Ini melibatkan banyak komponen dasar yang harus dipahami sebelum kita bisa membangun model yang lebih canggih. Beberapa komponen ini bisa sangat rumit dan menantang.

**Mengapa Penting?** Dengan NLP, kita bisa membuat komputer yang bisa memahami teks dan ucapan manusia, seperti chatbot, asisten virtual, dan sistem terjemahan otomatis. Ini membantu kita mengolah dan memahami informasi dalam jumlah besar dengan lebih efisien.

## Alur Kerja Pemrosesan Teks dalam NLP

Untuk memahami bagaimana memproses teks, penting untuk mengetahui alur kerja umum dalam Pemrosesan Bahasa Alami (NLP). Berikut adalah langkah-langkah dasarnya:



1. **Mengumpulkan Data Berlabel**: Langkah pertama adalah mengumpulkan data yang sudah diberi label. Model yang diawasi (supervised) atau semi-diawasi (semi-supervised) membutuhkan data ini untuk beroperasi. Misalnya, jika kita ingin membuat model yang bisa mengenali emosi dalam teks, kita butuh banyak contoh teks yang sudah diberi label emosi seperti "senang", "sedih", dll.

2. **Normalisasi dan Featurisasi Data**: Setelah data terkumpul, langkah berikutnya adalah menormalkan dan membuat fitur dari data tersebut. Model komputer sulit memproses teks mentah karena ada banyak struktur tersembunyi dalam teks yang perlu diungkapkan. Normalisasi bisa melibatkan mengubah semua huruf menjadi kecil, menghapus tanda baca, dll. Featurisasi adalah proses mengubah teks menjadi bentuk yang bisa dipahami oleh model, seperti angka atau vektor.

3. **Membangun Model**: Langkah terakhir adalah membangun model dengan input yang sudah diproses. Meskipun ada banyak model unik dalam NLP, di sini kita akan menggunakan jaringan saraf dalam sederhana dan fokus pada normalisasi dan featurisasi. 

Seringkali, tiga tahap terakhir ini berjalan dalam siklus, meskipun diagram mungkin terlihat linier. Dalam industri, menambahkan fitur baru membutuhkan lebih banyak usaha dan sumber daya untuk dijalankan. Oleh karena itu, penting bahwa fitur-fitur tersebut benar-benar menambah nilai.

Dengan pendekatan ini, kita akan menggunakan model sederhana untuk memvalidasi berbagai langkah normalisasi/vektorisasi/featurisasi.
