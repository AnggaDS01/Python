# **Ekstraksi Fitur dari Gambar**

Ekstraksi fitur adalah proses mengenali pola visual dalam gambar dan mengekstraksi fitur lokal yang dapat membedakan objek yang tidak dikenal. Sebelum melakukan ekstraksi fitur, penting untuk memahami histogram gambar. Histogram gambar adalah distribusi intensitas warna dari gambar tersebut.

#### Membuat Histogram Gambar

Berikut adalah kode Python yang digunakan untuk membuat histogram gambar mobil:

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

# Membuka gambar
image = Image.open('../car.png')
plt.imshow(image)

# Mengonversi gambar menjadi array numpy
image_arr = np.asarray(image)

# Menampilkan bentuk array gambar
print(image_arr.shape)

# Warna yang akan dianalisis
color = ('blue', 'green', 'red')

# Menghitung dan menampilkan histogram untuk setiap warna
for i, histcolor in enumerate(color):
    carhistogram = cv2.calcHist([image_arr], [i], None, [256], [0, 256])
    plt.plot(carhistogram, color=histcolor)
    plt.xlim([0, 256])

plt.show()
```

Kode di atas pertama-tama mengimpor pustaka Python yang diperlukan seperti `cv2` (OpenCV), `NumPy` (untuk perhitungan array), `PIL` (untuk mengimpor gambar), dan `Matplotlib` (untuk membuat grafik). Setelah itu, gambar dikonversi menjadi array dan histogram untuk setiap warna (R, G, dan B) dihitung dan ditampilkan.

#### Penjelasan Histogram

Histogram menunjukkan bahwa puncak intensitas warna R, G, dan B berada di sekitar 100, dengan puncak kedua di sekitar 150. Ini berarti bahwa rata-rata warna mobil adalah abu-abu. Frekuensi 0 pada nilai intensitas 200 menunjukkan bahwa mobil tersebut tidak berwarna putih. Demikian pula, frekuensi 0 pada nilai intensitas 50 menunjukkan bahwa gambar tersebut tidak sepenuhnya hitam.

### Contoh Aplikasi dan Perhitungan

#### Contoh Perhitungan

Misalkan kita memiliki dua gambar, A dan B, dan kita ingin membandingkan histogram mereka untuk menentukan seberapa mirip kedua gambar tersebut. Kita dapat menggunakan metode korelasi untuk membandingkan histogram:

```python
# Menghitung histogram untuk gambar A dan B
histA = cv2.calcHist([imageA], [0], None, [256], [0, 256])
histB = cv2.calcHist([imageB], [0], None, [256], [0, 256])

# Normalisasi histogram
cv2.normalize(histA, histA)
cv2.normalize(histB, histB)

# Menghitung korelasi antara histogram A dan B
correlation = cv2.compareHist(histA, histB, cv2.HISTCMP_CORREL)
print(f'Correlation: {correlation}')
```

Jika nilai korelasi mendekati 1, maka kedua gambar tersebut sangat mirip.

## **Pencocokan Gambar Menggunakan OpenCV**

Pencocokan gambar adalah teknik yang mencocokkan dua gambar berbeda untuk menemukan fitur umum. Teknik pencocokan gambar memiliki banyak aplikasi praktis, seperti mencocokkan sidik jari, mencocokkan warna karpet dengan warna lantai atau dinding, mencocokkan foto untuk menemukan dua gambar dari orang yang sama, atau membandingkan cacat manufaktur untuk mengelompokkannya ke dalam kategori yang serupa untuk analisis yang lebih cepat.

#### Teknik Pencocokan Gambar di OpenCV

Dua metode yang umum digunakan adalah BruteForce (BFMatcher) dan Fast Library for Approximate Nearest Neighbors (FLANN). Berikut adalah penjelasan singkat tentang kedua metode ini:

1. **BFMatcher**: Metode ini membandingkan jarak Hamming antara gambar uji dan setiap bagian dari gambar target untuk menemukan kecocokan terbaik.
2. **FLANN**: Metode ini lebih cepat tetapi hanya menemukan tetangga terdekat yang mendekatijadi, ia menemukan kecocokan yang baik, tetapi tidak selalu yang terbaik.

#### Aplikasi Pencocokan Gambar

1. **Pengenalan Sidik Jari**: Digunakan dalam sistem keamanan untuk mencocokkan sidik jari dengan database.
2. **Pencocokan Warna**: Membantu dalam desain interior untuk mencocokkan warna karpet dengan warna lantai atau dinding.
3. **Pengenalan Wajah**: Digunakan untuk mencocokkan foto untuk menemukan dua gambar dari orang yang sama.
4. **Analisis Cacat Manufaktur**: Membantu dalam mengelompokkan cacat manufaktur ke dalam kategori yang serupa untuk analisis yang lebih cepat.

#### Mengapa Kita Perlu Mempelajarinya?

Memahami dan menerapkan teknik pencocokan gambar sangat penting dalam berbagai bidang seperti keamanan, desain, dan manufaktur. Dengan kemampuan untuk mencocokkan gambar, kita dapat mengembangkan aplikasi yang lebih cerdas dan efisien, mulai dari sistem keamanan hingga aplikasi desain interior yang dapat membantu dalam pengambilan keputusan yang lebih baik.

````python
image_tile_rgb = Image.open(parent_path + 'tile.jpeg').convert('L')
image_tile_bgr = np.array(image_tile_rgb)

image_bathroom_rgb = Image.open(parent_path + 'bathroom_image.jpeg').convert('L')
image_bathroom_bgr = np.array(image_bathroom_rgb)


display_images(images=[image_tile_rgb, image_bathroom_bgr], is_gray=True)
````

````python
# https://www.geeksforgeeks.org/feature-extraction-and-image-classification-using-opencv/#conclusion

# https://github.com/opencv/opencv/tree/4.x/data

# # Inisialisasi SIFT, SIFT dapat menemukan titik tertentu dalam gambar dengan melihat ukuran, orientasi, dan detail lokalnya
# sift = cv2.SIFT_create()
# # Deteksi dan komputasi keypoints dan deskriptor
# kp1, des1 = sift.detectAndCompute(image_tile_bgr, None)
# kp2, des2 = sift.detectAndCompute(image_bathroom_bgr, None)

# # BFMatcher dengan jarak Hamming
# bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
# matches = bf.match(des1, des2)
# # Mengurutkan kecocokan berdasarkan jarak
# matches = sorted(matches, key=lambda x: x.distance)

# Inisialisasi ORB, ORB adalah alternatif SIFT yang lebih cepat, cocok untuk aplikasi real-time.
orb = cv2.ORB_create()
# Deteksi dan komputasi keypoints dan deskriptor
kp1,des1 = orb.detectAndCompute(image_tile_bgr, None)
kp2,des2 = orb.detectAndCompute(image_bathroom_bgr, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1,des2)
matches = sorted(matches,key=lambda x:x.distance)

floor_matches = cv2.drawMatches(
    image_tile_bgr, 
    kp1,
    image_bathroom_bgr,
    kp2,
    matches[:10], 
    None, 
    flags=2
)
````

````python
display_images(images=[floor_matches], figsize_per_image=(8,4))
````

````python
sift = cv2.SIFT_create()
kp1,des1 = sift.detectAndCompute(image_tile_bgr, None)
kp2,des2 = sift.detectAndCompute(image_bathroom_bgr, None)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
````

````python
good = []
for match1,match2 in matches:
    if match1.distance < 1.2*match2.distance:
        good.append([match1])
````

````python
draw_params = dict(matchColor=(0,0,255),singlePointColor=(0,0,255))
sift_matches = cv2.drawMatchesKnn(image_tile_bgr, kp1, image_bathroom_bgr, kp2,good, None, **draw_params)
````

````python
display_images(images=[sift_matches], figsize_per_image=(8,4))
````

````python
sift = cv2.SIFT_create()
kp1,des1 = sift.detectAndCompute(image_tile_bgr, None)
kp2,des2 = sift.detectAndCompute(image_bathroom_bgr, None)

flann_index_kdtree=0
index_params = dict(algorithm = flann_index_kdtree,trees=7)
search_params = dict(checks=50)

flann = cv2.FlannBasedMatcher(index_params,search_params)
matches = flann.knnMatch(des1,des2,k=2)
````

````python
matchesMask = [[0,0] for i in range(len(matches))]
for i,(match1,match2) in enumerate(matches):
    if match1.distance <1.2*match2.distance:
        matchesMask[i]=[1,0]
````

````python
draw_params = dict(matchColor=(255,0,0),singlePointColor=(0,0,255),matchesMask=matchesMask,flags=2)
flann_matches = cv2.drawMatchesKnn(image_tile_bgr,kp1,image_bathroom_bgr,kp2,matches,None,**draw_params)
````

````python
display_images(images=[flann_matches], figsize_per_image=(10,5))
````
