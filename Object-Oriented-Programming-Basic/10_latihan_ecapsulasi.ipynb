```python
class Hero:
  __get_arguments = []

  def __init__(self, input_name, input_health, input_armor, input_power):
    self.__name = input_name
    self.__health_base = input_health
    self.__armor_base = input_armor
    self.__power_base = input_power
    self.__exp_base = 0
    self.__level = 1

    self.__health_max = self.__health_base
    self.__exp_max = 100

    Hero.__get_arguments.append(self)

  def attack(self, enemy):
    self.set_exp(self.__power_base - ( enemy.__armor_base * .25), enemy.__health_base)
    enemy.__health_base -= self.__power_base
    if enemy.__health_base <= 0:
      enemy.__health_base = 0
      raise Exception(f'{enemy.__name} died, you cannot attack again')
    print(f'{self.__name} menyerang {enemy.__name} level {enemy.__level} sisa nyawa: {enemy.__health_base}/{enemy.__health_max}')

  def set_exp(self, power, enemy_health):
    if enemy_health > 0:
      self.__exp_base += power
      if (self.__exp_base >= self.__exp_max ):
        print(f'{self.__name} level up')
        self.__level += 1
        self.__exp_base -= self.__exp_max
        self.__exp_max = self.__level * 100

        self.__health_max = round(self.__health_max + self.__level / .25, 2)
        self.__power_base = round(self.__power_base + self.__level / .45, 2)
        self.__armor_base = round(self.__armor_base + self.__level / .5, 2)

  @classmethod
  def get_arguments(cls):
    return cls.__get_arguments

  @property
  def get_info(self):
    return (f'{self.__name} level {self.__level} (exp = {round(self.__exp_base, 2)}/{self.__exp_max}):\n\thealth: {self.__health_base}/{self.__health_max}\n\tarmor: {self.__armor_base}\n\tpower: {self.__power_base}')

  def __repr__(self):
    return (f'Hero({self.__name}, {self.__health_base}, {self.__armor_base}, {self.__power_base})')

knight = Hero('Knight', 700, 50, 75)
archer = Hero('archer', 400, 25, 100)

print(knight.get_info)
knight.attack(archer)
print(knight.get_info)
knight.attack(archer)
print(knight.get_info)

print('\n')
archer.attack(knight)
print(archer.get_info)
archer.attack(knight)
print(archer.get_info)
archer.attack(knight)
print(archer.get_info)
archer.attack(knight)
print(archer.get_info)

print('\n')
knight.attack(archer)
print(knight.get_info)
knight.attack(archer)
print(knight.get_info)
```
